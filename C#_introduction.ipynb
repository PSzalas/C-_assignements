> Przeciąż metodę `ToString()`, która ma zawierać zestaw podstawowych informacji. Zadanie można zrealizować na dwa sposoby: 
1. metody do obliczeń pola i obwodu zaimplementować w klasie abstrakcyjnej używając wzoru:

$
	P =\sqrt{p\cdot(p-A)\cdot(p-B)\cdot(p-C)}
$

gdzie: $p$ to połowa obwodu trójkąta. Symbole $A$, $B$, $C$ to długości poszczególnych boków. Oznaczyć metody obliczające pole i obwód jako abstrakcyjne i przeciążyć je w klasie szczegółowej.

```polyglot-notebook
//Zadanie 1 Wykonali: Bartosz, Patryk S

public abstract class Trojkat
{
    public double bokA;
    public double bokB;
    public double bokC;

    public double obwod()
    {
        return bokA + bokB + bokC;
    }
    public double pole()
    {
        return Math.Sqrt((this.obwod() / 2) * (this.obwod() / 2 - bokA) * (this.obwod() / 2 - bokB) * (this.obwod() / 2 - bokC));
    }

    public override string ToString()
    {
        return "BokA: " + this.bokA + " BokB: " + this.bokB + " bokC: " + this.bokC + " Obwod: " + this.obwod() + " Pole: " + this.pole();
    }


}
public class Rownobok : Trojkat
{
    public Rownobok(double bok)
    {
        this.bokA = bok;
        this.bokB = bok;
        this.bokC = bok;
    }
}

public class Rownoramien : Trojkat
{


    public Rownoramien(double bokA, double bokB)
    {
        this.bokA = bokA;
        this.bokB = bokA;
        this.bokC = bokB;
    }
}
public class Prostok : Trojkat
{
    public Prostok(double bokA, double bokB)
    {
        this.bokA = bokA;
        this.bokB = bokB;
        this.bokC = Math.Sqrt(this.bokA * this.bokA + this.bokB * this.bokB);
    }

}

class Program
{
    static void Main(string[] args)
    {
        Rownobok rownobok = new Rownobok(2);
        Console.WriteLine(rownobok.ToString());
        Rownoramien rownoramieny = new Rownoramien(2, 3);
        Console.WriteLine(rownoramieny.ToString());
        Prostok prostokatny = new Prostok(2, 3);
        Console.WriteLine(prostokatny.ToString());
    }
}
```

```polyglot-notebook
//Zadanie 2,3 i 4 Wykonali: Bartosz, Patryk S

using System.Configuration;
using System;
using System.IO;

public class Employee<T>
{
    public T Id { get; set; }
    public string FirstName { get; set; }
    public string LastName { get; set; }
    public string Position { get; set; }
    public string Department { get; set; }
    public string PhoneNumber { get; set; }

    public Employee(T id, string firstName, string lastName, string position, string department, string phoneNumber)
    {
        Id = id;
        FirstName = firstName;
        LastName = lastName;
        Position = position;
        Department = department;
        PhoneNumber = phoneNumber;
    }

    public bool Validate()
    {
        bool isValid = true;

        if (string.IsNullOrEmpty(FirstName))
        {
            throw new ArgumentException("Imię nie może być puste.");
            isValid = false;
        }
        if (string.IsNullOrEmpty(LastName))
        {
            throw new ArgumentException("Nazwisko nie może być puste.");
            isValid = false;
        }
        if (string.IsNullOrEmpty(Position))
        {
            throw new ArgumentException("Stanowisko nie może być puste.");
            isValid = false;
        }
        if (string.IsNullOrEmpty(Department))
        {
            throw new ArgumentException("Departament nie może być pusty.");
            isValid = false;
        }
        if (string.IsNullOrEmpty(PhoneNumber))
        {
            throw new ArgumentException("Numer telefonu nie może być pusty.");
            isValid = false;
        }

        return isValid;
    }

    public void Show()
    {
        Console.WriteLine(Format());
    }

    public bool IsMatch(Employee<T> employee)
    {
        return Id.Equals(employee.Id) &&
            FirstName == employee.FirstName &&
            LastName == employee.LastName &&
            Position == employee.Position &&
            Department == employee.Department &&
            PhoneNumber == employee.PhoneNumber;
    }

    private string Format()
    {
        return $"\nId: {Id}\nImię: {FirstName}\nNazwisko: {LastName}\nStanowisko: {Position}\nDepartament: {Department}\nNumer telefonu: {PhoneNumber}";
    }


    public void Encrypt()
    {
        int blockSize = int.Parse(ConfigurationManager.AppSettings["BlockSize"]);

        Id = (T)(object)(int)((int)(object)Id + blockSize);

        FirstName = EncryptString(FirstName, blockSize);
        LastName = EncryptString(LastName, blockSize);
        Position = EncryptString(Position, blockSize);
        Department = EncryptString(Department, blockSize);
        PhoneNumber = EncryptString(PhoneNumber, blockSize);
    }

    public void Decrypt()
    {
        int blockSize = int.Parse(ConfigurationManager.AppSettings["BlockSize"]);

        Id = (T)(object)(int)((int)(object)Id - blockSize);

        FirstName = DecryptString(FirstName, blockSize);
        LastName = DecryptString(LastName, blockSize);
        Position = DecryptString(Position, blockSize);
        Department = DecryptString(Department, blockSize);
        PhoneNumber = DecryptString(PhoneNumber, blockSize);
    }

    private string EncryptString(string str, int blockSize)
    {

        char[] chars = str.ToCharArray();

        for (int i = 0; i < chars.Length; i++)
        {
            char c = chars[i];
            if (c >= 'a' && c <= 'z')
            {
                c = (char)('a' + ((c - 'a' + blockSize) % 26));
            }
            else if (c >= 'A' && c <= 'Z')
            {
                c = (char)('A' + ((c - 'A' + blockSize) % 26));
            }
            else if (c >= '0' && c <= '9')
            {
                c = (char)('0' + ((c - '0' + blockSize) % 10));
            }

            chars[i] = c;
        }
        return new string(chars);
    }

    private string DecryptString(string str, int blockSize)
    {

        char[] chars = str.ToCharArray();

        for (int i = 0; i < chars.Length; i++)
        {
            char c = chars[i];
            if (c >= 'a' && c <= 'z')
            {
                c = (char)('a' + ((c - 'a' - blockSize + 26) % 26));
            }
            else if (c >= 'A' && c <= 'Z')
            {
                c = (char)('A' + ((c - 'A' - blockSize + 26) % 26));
            }
            else if (c >= '0' && c <= '9')
            {
                c = (char)('0' + ((c - '0' - blockSize + 10) % 10));
            }

            chars[i] = c;
        }
        return new string(chars);
    }
}

public class EmployeeDirectory<T>
{
    private List<Employee<T>> _employees;

    public EmployeeDirectory()
    {
        _employees = new List<Employee<T>>();
    }

    public void Add(Employee<T> employee)
    {
        if (employee.Validate())
        {
            _employees.Add(employee);
        }
    }

    public void Remove(Employee<T> employee)
    {
        _employees.Remove(employee);
    }

    public void Show()
    {
        foreach (Employee<T> employee in _employees)
        {
            employee.Show();
        }
    }

    public Employee<T> Search(Employee<T> employee)
    {
        foreach (Employee<T> e in _employees)
        {
            if (e.IsMatch(employee))
            {
                return e;
            }
        }

        return null;
    }

    public void ReadFromTxt(string filePath)
    {
        using (StreamReader reader = new StreamReader(filePath))
        {
            string line;
            while ((line = reader.ReadLine()) != null)
            {
                EmployeeBuilder<T> builder = new EmployeeBuilder<T>();
                string[] parts = line.Split('|');
                T id = default(T);
                if (typeof(T) == typeof(int))
                {
                    id = (T)(object)int.Parse(parts[0]);
                }
                else if (typeof(T) == typeof(string))
                {
                    id = (T)(object)parts[0];
                }
                builder.SetId(id);
                builder.SetFirstName(parts[1]);
                builder.SetLastName(parts[2]);
                builder.SetPosition(parts[3]);
                builder.SetDepartment(parts[4]);
                builder.SetPhoneNumber(parts[5]);

                Employee<T> employee = builder.Build();
                Add(employee);
            }
        }
    }

    public void WriteToTxt(string filePath)
    {
        using (StreamWriter writer = new StreamWriter(filePath))
        {
            foreach (Employee<T> employee in _employees)
            {
                string line = $"{employee.Id}|{employee.FirstName}|{employee.LastName}|{employee.Position}|{employee.Department}|{employee.PhoneNumber}";
                writer.WriteLine(line);
            }
        }
    }

    public void Encrypt()
    {
        foreach (Employee<T> employee in _employees)
        {
            employee.Encrypt();
        }
    }

    public void Decrypt()
    {
        foreach (Employee<T> employee in _employees)
        {
            employee.Decrypt();
        }
    }
}

public class EmployeeBuilder<T>
{
    private T _id;
    private string _firstName;
    private string _lastName;
    private string _position;
    private string _department;
    private string _phoneNumber;

    public void SetId(T id)
    {
        _id = id;
    }

    public void SetFirstName(string firstName)
    {
        _firstName = firstName;
    }

    public void SetLastName(string lastName)
    {
        _lastName = lastName;
    }

    public void SetPosition(string position)
    {
        _position = position;
    }

    public void SetDepartment(string department)
    {
        _department = department;
    }

    public void SetPhoneNumber(string phoneNumber)
    {
        _phoneNumber = phoneNumber;
    }

    public Employee<T> Build()
    {
        return new Employee<T>(_id, _firstName, _lastName, _position, _department, _phoneNumber);
    }
}

public class Program
{
    public static void Main(string[] args)
    {
        EmployeeDirectory<int> directory = new EmployeeDirectory<int>();

        Employee<int> employee1 = new Employee<int>(1, "Franek", "Dolas", "Kierownik", "Sprzedaży", "234689326");
        Employee<int> employee2 = new Employee<int>(2, "Patryk", "S", "Programista", "IT", "967439562");


        directory.Add(employee1);
        directory.Add(employee2);
        directory.Show();

        Employee<int> employee3 = new Employee<int>(2, "Patryk", "S", "Programista", "IT", "967439562");
        Employee<int> result = directory.Search(employee3);
        if (result != null)
        {
            Console.WriteLine("\nPracownik znaleziony:");
            result.Show();
        }
        else
        {
            Console.WriteLine("\nPracownik nie znaleziony.");
        }

        string blockSizeString = ConfigurationManager.AppSettings["BlockSize"];
        Console.WriteLine(blockSizeString);

        Console.WriteLine("Orginalne dane:");
        directory.Show();


        directory.Encrypt();

        Console.WriteLine("\nZakryptowane dane:");
        directory.Show();

        directory.Decrypt();


        Console.WriteLine("\nOdkryptowane dane:");
        directory.Show();
    }
}

 
```
